ARIMA es un modelo estadstico de series temporales, que tiene una estructura propia para tratar con datos dependientes del tiempo.

ARIMA est diseado para trabajar con datos de series temporales, donde los valores observados en el tiempo presente dependen de los valores pasados (y en algunos casos, de la tendencia o estacionalidad).

Regresin lineal mltiple asume que las observaciones son independientes entre s (no se toma en cuenta la dependencia temporal entre los datos).

Parmetros en ARIMA o SARIMA (Seasonal ARIMA) (p, d, q) + (P, D, Q)

Parmetro p: Captura la relacin de un valor con sus propios valores pasados para ver la dependencia temporal. Este coeficiente mide el numero de rezagos (lags) que tiene el valor y que son explicados por observaciones anteriores. 

Parmetro d: Mide como de constantes sn la varianza y la media dentro de una serie temporal. Muestra el nmero de diferenciaciones. 

Parmetro q: Muestra la Media mvil, que s la relacin entre un valor y los errores del pasado, mostrando el nmero de errores pasados considerados.

IMPORTANTE: ARIMA no es un modelo de regresin lineal mltiple porque se basa en los valores pasados de la serie temporal y en los errores pasados, mientras que la regresin lineal mltiple usa variables independientes (predictoras) para estimar la variable dependiente.

Los coeficientes son estimados para representar las relaciones autoregresivas y de media mvil en el tiempo. s entonces, un modelo de dependencia temporal, que captura patrones en los datos a lo largo del tiempo (tendencias, ciclos, estacionalidad) y predice los valores futuros en funcin de estos patrones.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Preparacin de los datos
california_base['Current Year Week Ending'] = pd.to_datetime(california_base['Current Year Week Ending'])
price_series = california_base[['Current Year Week Ending', 'ASP Current Year']].set_index('Current Year Week Ending')

# 2. Verificar estacionariedad con la prueba de Dickey-Fuller
result = adfuller(price_series['ASP Current Year'])
print(f"Estadstico ADF: {result[0]}")
print(f"Valor p: {result[1]}")

# Si el valor p es mayor que 0.05, la serie no es estacionaria y deberamos diferenciarla
if result[1] > 0.05:
    price_series['ASP Current Year'] = price_series['ASP Current Year'].diff().dropna()

# 3. Identificacin de los parmetros (p, d, q) usando ACF y PACF
plot_acf(price_series['ASP Current Year'].dropna())
plot_pacf(price_series['ASP Current Year'].dropna())
plt.show()

# A continuacin, se debe analizar visualmente los grficos ACF y PACF para decidir los valores de p, q, y d.
# Aqu asumimos que hemos identificado estos parmetros de los grficos.

# Asumamos que los parmetros seleccionados despus de observar ACF y PACF son:
p = 1  # Ejemplo: Nmero de lags en el modelo AR (autoregresivo)
d = 1  # Ejemplo: Nmero de diferenciaciones necesarias para hacer la serie estacionaria
q = 1  # Ejemplo: Nmero de lags de los errores en el modelo MA (media mvil)

# 4. Ajuste del modelo SARIMA (si la estacionalidad es importante)
# Si se detecta estacionalidad semanal, asumimos que el perodo es de 52 semanas.
# Los parmetros estacionales P, D, Q se determinan de manera similar a los anteriores.

P = 1  # Estacionalidad - autoregresivo (estacional)
D = 1  # Estacionalidad - diferenciacin
Q = 1  # Estacionalidad - media mvil

model = SARIMAX(price_series['ASP Current Year'], 
                order=(p, d, q),  # Parmetros de ARIMA
                seasonal_order=(P, D, Q, 52),  # Estacionalidad semanal (52 semanas)
                enforce_stationarity=False, 
                enforce_invertibility=False)
results = model.fit()

# 5. Evaluar el modelo
print(results.summary())
```

Primeramente, podemos confirmar que el modelo es estacionario, es decir, que la media, la varianza y la autocorrelacin se mantienen a lo largo de la serie temporal. Esto se confirma gracias a que el parmetro ADF (Dickey-Fuller) nos arroja un p-valor de 0,05 lo que nos hace  rechazar la hiptesis nula (por muy poco) y nos afirma que hay un componente de estacionariedad en este modelo. 

Hemos utilizado concretamente un modelo SARIMAX (1, 1, 1) x (1 , 1, 1, 52) quiere decir que los valores de p, q y d son 1, y como tenemos el componente estacional semanal de 52 semanas, el ltimo coeficiente s 52. 
Por lo tanto:
p = 1, implica que se usa 1 rezago autoregresivo en el componente AR (esto es, el valor de la serie temporal pasada se utiliza para predecir el valor actual).

d = 1, implica que se realiza una diferenciacin de primer orden para hacer que la serie sea estacionaria.

q = 1, implica que se usa 1 rezago de los residuos pasados en el componente de media mvil.

```python
# Generamos las predicciones para los prximos 10 periodos
forecast = results.get_forecast(steps=10)  # Predecir los prximos 10 periodos
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# Aseguramos que las fechas de las predicciones estn alineadas con las fechas de los datos histricos
last_date = price_series.index[-1]  # ltima fecha de los datos histricos
forecast_dates = pd.date_range(start=last_date, periods=11, freq='W')[1:]  # 10 pasos hacia adelante (semana a semana)

# Ajustamos el ndice de las predicciones para que tenga las fechas correctas
forecast_mean.index = forecast_dates
forecast_ci.index = forecast_dates

# Graficar la prediccin
plt.figure(figsize=(10, 5))
plt.plot(price_series.index, price_series['ASP Current Year'], label='Datos histricos')
plt.plot(forecast_mean.index, forecast_mean, color='red', label='Prediccin')
plt.fill_between(forecast_ci.index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='pink', alpha=0.3)
plt.legend()
plt.show()
```

```python
# Obtener los residuos del modelo
residuals = results.resid

# Graficar los residuos
plt.figure(figsize=(10, 5))
plt.plot(residuals)
plt.title('Grfico de residuos')
plt.xlabel('Tiempo')
plt.ylabel('Residuo')
plt.show()

# Grfico de histograma de residuos
plt.figure(figsize=(10, 5))
plt.hist(residuals, bins=30, edgecolor='black')
plt.title('Distribucin de los residuos')
plt.xlabel('Residuo')
plt.ylabel('Frecuencia')
plt.show()
```

Este grfico nos muestra que los residuos no se comportan de manera estandarizada, s decir que no siguen un ruido de camino blanco. Por lo tanto el modelo no padece de sobreajuste y puede ser til para intentar predecir errores futuros tambin.

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Usaremos los precios de los aguacates, que en tu caso parece que estn en la columna 'ASP Current Year'
series_temporal = california_base['ASP Current Year']  # Ajusta el nombre de la columna si es necesario

# Graficar la FAS (Funcin de Autocorrelacin) de la serie temporal
plt.figure(figsize=(12, 6))

# FAS (ACF)
plt.subplot(1, 2, 1)  # Primer grfico de la fila 1, columna 2
plot_acf(series_temporal, lags=40, ax=plt.gca())  # FAS con lags hasta el 40
plt.title('Funcin de Autocorrelacin (FAS) de la Serie Temporal')

# FAP (PACF)
plt.subplot(1, 2, 2)  # Segundo grfico de la fila 1, columna 2
plot_pacf(series_temporal, lags=40, ax=plt.gca())  # FAP con lags hasta el 40
plt.title('Funcin de Autocorrelacin Parcial (FAP) de la Serie Temporal')

# Ajuste para evitar solapamientos
plt.tight_layout()
plt.show()
```

La FAS i la FAP, nos devuelven que claramente se evidencia una tendencia estacional, ya que los rezagos que hay con las temporalidades del periodo t, vienen correlacionadas desde varios periodos atrs.
