```python
import pandas as pd
import matplotlib.pyplot as plt

filtered_df_regions['Current Year Week Ending'] = pd.to_datetime(filtered_df_regions['Current Year Week Ending'])

filtered_df_regions['Quarter'] = filtered_df_regions['Current Year Week Ending'].dt.to_period('Q')
quarterly_mean = filtered_df_regions.groupby('Quarter')['Total Bulk and Bags Units'].mean()

plt.figure(figsize=(10, 6))
quarterly_mean.plot(kind='bar')
plt.xlabel('Trimestre')
plt.ylabel('Media del total de unidades a granel y en sacos (lb)')
plt.title('Media del total de unidades a granel y en sacos (lb) por Trimestre')
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

filtered_df_regions["Current Year Week Ending"] = pd.to_datetime(filtered_df_regions["Current Year Week Ending"])

# Extract the quarter from the 'Current Year Week Ending' column
filtered_df_regions['Quarter'] = filtered_df_regions['Current Year Week Ending'].dt.to_period('Q')

# Group by quarter and calculate the mean of 'Total Bulk and Bags Units'
quarterly_mean = filtered_df_regions.groupby('Quarter')['Total Bulk and Bags Units'].mean().reset_index()

quarterly_mean.head()
```

```python

# Plot the data
fig, ax1 = plt.subplots()

# Plot the 'Current Year Week Ending' by quarter
ax1.plot(filtered_df_regions['Current Year Week Ending'], filtered_df_regions['Quarter'], 'b-')
ax1.set_xlabel('Current Year Week Ending')
ax1.set_ylabel('Quarter', color='b')

# Create another y-axis for the mean of 'Total Bulk and Bags Units'
ax2 = ax1.twinx()
ax2.plot(quarterly_mean['Quarter'].astype(str), quarterly_mean['Total Bulk and Bags Units'], 'r-')
ax2.set_ylabel('Mean of Total Bulk and Bags Units', color='r')

plt.title('Current Year Week Ending by Quarter and Mean of Total Bulk and Bags Units')
plt.show()
```
