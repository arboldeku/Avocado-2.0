Tuvimos la problemtica de que al querer mezclar la base de datos de Suppliers y la que contenia los datos de Consumo, no podiamos ignorar que la de Consumo tena dos entradas (Una para orgnico y otra para convencional), mientras que la de Suppliers tenia solamente la entrada de los datos con las cifras de exportacin total desde California.

Dada esta problemtica nos decidimos por calcular una columna que recogiera el porcentaje de Exportaciones de Californio respecto al Consumo total en California. A pesar de ello esta medida es completamente inconsistente ya que el consumo de California puede venir de cualquier otro del os proveedores. Es por eso que decidimos calcular un Coeficiente del peso del consumo de aguacates en california respecto al resto de regiones. Una vez encontrado este peso, si fuera necesario podramos implementar escala logartmica para normalizar los datos. 

```python
def calcular_pesos_regionales_y_consumo(df, regiones):
    """
    Calcula los pesos normalizados por regin y determina el consumo de California.

    Args:
    - df: DataFrame con los datos de consumo por regin.
    - regiones: Lista de las regiones a considerar.

    Returns:
    - pesos_regionales: Diccionario con los pesos normalizados por regin.
    - consumo_california: Consumo total ponderado de California.
    """
    # Inicializar el diccionario de pesos regionales
    pesos_regionales = {}

    # Sumar el consumo total de todas las regiones
    total_consumo_global = df['Total Unidades a Granel y Bolsas'].sum()

    # Iterar sobre las regiones para calcular el peso de cada una
    for region in regiones:
        # Sumar el consumo total de la regin
        consumo_region = df.loc[
            df['Geography'] == region, 'Total Unidades a Granel y Bolsas'
        ].sum()

        # Calcular el peso relativo de la regin
        peso_region = consumo_region / total_consumo_global if total_consumo_global > 0 else 0
        pesos_regionales[region] = peso_region

    # Verificar que la suma de los pesos sea igual a 1
    suma_pesos = sum(pesos_regionales.values())
    if round(suma_pesos, 5) == 1:
        print(f"La suma de los pesos es correcta: {suma_pesos}")
    else:
        print(f"Advertencia: La suma de los pesos es {suma_pesos}, no igual a 1")

    # Obtener el consumo de California
    consumo_california = total_consumo_global * pesos_regionales.get('California', 0)

    return pesos_regionales, consumo_california


# Regiones relevantes
regiones = ['California', 'Great Lakes', 'Midsouth', 'Northeast', 'Plains', 
            'South Central', 'Southeast', 'West']

# Calcular los pesos y el consumo de California
pesos_regionales, consumo_california = calcular_pesos_regionales_y_consumo(
    filtered_df_greaterregions, regiones
)

# Mostrar los resultados
print("Pesos regionales:")
for region, peso in pesos_regionales.items():
    print(f"{region}: {peso:.4f}")

print(f"\nConsumo ponderado de California: {consumo_california:.2f} unidades")
```

```python
def calcular_tasa_consumo_interno(df, peso_california):
    """
    Calcula la tasa de consumo interno en California para todos los registros del dataframe.

    Args:
    - df: DataFrame con los datos de California.
    - peso_california: Peso de California (0.1666).
    
    Modifica el DataFrame aadiendo una nueva columna 'tasa_consumo_interno_california'.
    """
    # Calcular la tasa de consumo interno en California usando el peso de California
    df['tasa_consumo_interno_california'] = df['California Supplier'] / (df['Total Unidades en Granel y Bolsas'] * peso_california)

    return df

# Definir el peso de California
peso_california = 0.1666

# Calcular la tasa de consumo interno en California para el dataframe
california_consumption = calcular_tasa_consumo_interno(california_consumption, peso_california)

# Mostrar las primeras filas del dataframe con la nueva columna
print(california_consumption[['Geography', 'California Supplier', 'Total Unidades en Granel y Bolsas', 'tasa_consumo_interno_california']].head())
```

Para encontrar la tasa de consumo local aproximada de California, hemos calculado un coeficiente del peso de California respecto al consumo total en TotalUS. Esto nos garantiza que ahora los valores estn ms ajustados a la realidad de que; Por los aguacates que son exportados de California, con un peso del 9% son consumidos en California, obteniendo as la tasa local de consumo de la siguiente manera. **Tasa de Consumo Local de California** = Exportaciones / (Consumo Total de California * Coeficiente peso relativo). Ahora ya le podemos dar sentido a los datos scrappeados y podemos proseguir con el modelo de regresin lineal.

Los resultados del modelo MCO son preocupantes porque los valores de R2 para ambos tipos de aguacate son bajos, y en el caso del orgnico, incluso negativos. Esto sugiere que el modelo actual no est capturando adecuadamente la relacin entre las variables independientes y el precio del aguacate.

Al tener datos con frecuencias muy distintas (datos en millones y datos en valores ms pequeos), nos disponemos a hacer un modelo LinLog, es decir, aplicar escala logartmica a aquellos datos que tengan dimensiones en miles o millones. Razones  de este ajuste moderado:
1. Relaciones no lineales: Es posible que las variables independientes tengan una relacin no lineal con el precio, lo que un modelo lineal simple no puede capturar.
2. Multicolinealidad: Las variables independientes podran estar correlacionadas entre s, lo que puede afectar la interpretacin y efectividad del modelo.
3. Datos insuficientes o ruido: Los datos de consumo local y elasticidad, que son esenciales para modelar precios, no estn incluidos o contienen ruido que dificulta obtener relaciones claras.
4. Distribucin desigual de los datos: Si algunas variables tienen valores extremos o estn sesgadas, esto puede afectar negativamente la performance del modelo.
